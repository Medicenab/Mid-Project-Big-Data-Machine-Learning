```python
class Vector:
    def __init__(self, point, origin=None):
        if origin == None:
            origin = [0]*len(point)
        self.origin = np.array(origin)
        self.point = np.array(point)
    def plot(self):
        plt.arrow(*self.origin, *(self.point-self.origin), head_width=.2, length_includes_head=True)
        
    def __add__(self, other):
        return Vector(self.point+other.point)
    
    def __sub__(self, other):
        return Vector(self.point-other.point)
    #def __mul__
    
    def norm(self):
        return np.sqrt(sum((self.point-self.origin)**2))
    
    def __repr__(self):
        return f"Vector -> {self.origin} {self.point}"
```

```python
Graph(title="Vector in Space")
for point in points:
    plt.scatter(*point)
v = Vector((5,7), (2,5))
v.plot()
```

```python
class Retangle:
    def __init__(self, origin, dx, dy):
        self.origin = np.array(origin)
        self.points = [self.origin, self.origin + (dx,0), self.origin + (dx,dy), self.origin + (0,dy)]
        self.coord = list(zip(*[*self.points, self.points[0]]))
    def draw(self):
        plt.plot(*self.coord)
    def fill(self, alpha=1, **kwargs):
        print(self.coord)
        plt.fill_between(*self.coord, alpha=alpha, **kwargs)
        
class Square(Retangle):
    def __init__(self,origin,l):
        super().__init__(origin, l, l)
```

```python
Graph(title="Squares", figsize=(6,6))
sq1 = Square((0,0), 3)
sq2 = Square((0,0), -3)
for s, c in zip([sq1, sq2],["lightblue","lightblue"]):
    s.fill(alpha=.3, color=c)
```

Como secta, tenían muchas reglas sobre la vida, como vivir y la filosofia. Ellos creían que todos los números deberían ser posibles de ser representados racionalmente, como una fracción. ### $\sqrt{2}$

No mucho querido por los pitagóricos, el numero $\sqrt{2}$ tiene algunas propriedades muy interesantes. 

El patrón ISO 216 define algunas reglas sobre el tamaño de los folios conocidos como A0, A1, A2, etc. Tal cómo fué concebido, el papel de tamaño A0 debería tener un área de $1m^2$. ¿Parece tarea bastante sencilla, no? Eso se puede conseguir con un folio que tenga 1m x 1m.

Entretanto, un folio cuadrado cuando dobrado a la mitad pasa a tener mitad del àrea, pero no mantiene las mismas proporciones. Un folio cuadrado dobrado a la mitad deja de ser cuadrado.

Resulta que la única proporción capaz de mantener el mismo `aspect ratio` cada vez que dobrado a la mitad es la proporción [$\sqrt{2}$](https://en.wikipedia.org/wiki/Square_root_of_2):1. Apenas una de las propriedades que torna ese numero importante en las ciencias y tecnologia.

Podemos pensar en las coordenadas de un punto $P(x,y)$ como la representación de un punto en concreto de una función matemática, donde $y=f(x)$, y por lo tanto $P=(x,f(x))$.

La ecuación que describe esa función `f` será la responsable por definir la trayectória que esa curva produce sobre el espacio. En el caso de una recta, la ecuación que la describe es:

$$
y = m*x + b
$$

En el caso del ejemplo arriba, podemos representar los dos puntos como:

$$
\begin{cases}
7*m + b = 4\\
3*m + b = 5
\end{cases}
$$

Convertiendo a la forma matricial, llegamos a lo siguiente:

$$
\begin{bmatrix} 7 & 1 \\ 3 & 1 \end{bmatrix} 
\times 
\left[ \begin{array}{c} m \\ b \end{array} \right]
=
\left[ \begin{array}{c} 4 \\ 5 \end{array} \right]
$$

Podemos utilizar de numpy para resolver el sistema y sacar los valores para `m` y `b`.

```python
# A·x = b

A = [[7,1],
     [3,1]]

b = [4,5]

m,b = np.linalg.solve(A,b)
m,b
```

¿Y si buscaramos una parábola?

La ecuación de la parábola $y=a*x^2 + b*x + c$ seria la ecuación que describiria la curva., por lo tanto tendríamos el siguiente sistema de ecuaciones:


$$
\begin{cases}
49*a + 7*b + c = 4\\
9*a + 3*b + c = 5
\end{cases}
$$

O en su forma matricial:

$$
\begin{bmatrix} 49 & 7 & 1 \\ 9 & 3 & 1 \end{bmatrix} 
\times 
\left[ \begin{array}{c} a \\ b \\ c \end{array} \right]
=
\left[ \begin{array}{c} 4 \\ 5 \end{array} \right]
$$

Ese sistema de ecuaciones no se puede resolver, pues la matrix A no es cuadrada.

```python
A = [[49,7,1],
     [9,3,1]]
b = [4,5]

a,b,c = np.linalg.solve(A,b)
```

No tenemos puntos suficientes para encontrar una solución para ese sistema de ecuaciones. Seria necesario tener un punto más, por ejemplo P(5,0).

$$
\begin{cases}
49*a + 7*b + c = 4\\
9*a + 3*b + c = 5\\
25*a + 5*b + c = 0
\end{cases}
$$

```python
A = [[49,7,1],
     [9,3,1],
     [25, 5, 1]]
b = [4,5,0]

a,b,c = np.linalg.solve(A,b)
a,b,c
```

```python
Graph(title="Parabola crossing two points", figsize=(6,6))
P1 = (7,4)
P2 = (3,5)
P3 = (5,0)
for p in [P1,P2,P3]:
    plt.scatter(*p)
    
parabola = lambda x: a*x**2 + b*x + c
# np.arange(start, stop, step)
x = np.arange(0,8,.1)
y = parabola(x)

plt.plot(x,y)
```

```python
Graph(title="Area under a curve", figsize=(12,6))
x = np.arange(0,3,.01)
curve = lambda x : 3*np.cos(x) + .1*x + 5
y = curve(x)
plt.plot(x,y, color="red")

n = 30
maxi = max(x)
step = len(x)//n

for i in range(n):
    Retangle((maxi/n*i,0), maxi/n, min(y[step*i:step*i+step-1])).fill()
```

```python
Graph(title="Area under a curve", figsize=(12,6))
x = np.arange(0,3,.01)
curve = lambda x : 3*np.cos(x) + .1*x + 5
y = curve(x)
plt.plot(x,y, color="red")

dx = 0.01
a,b = 0,3

area = 0
for i in range(int((b-a)/dx)):
    area += dx*curve(dx*i+dx)
    Retangle((dx*i,0), dx, curve(dx*i+dx)).fill()
plt.annotate(f"Area: {area}",(2,7));
```

```python
x = np.arange(0,3,.01)
curve = lambda x : 3*np.cos(x) + .1*x + 5
y = curve(x)

dx_list = np.arange(2,0,-.0001)
area_list = []

for dx in dx_list:
    a,b = 0,3
    area = 0
    for i in range(int((b-a)/dx)):
        area += dx * curve(dx*i+dx)
    area_list.append(area)

Graph(title="Integration", figsize=(12,6))
plt.plot(-dx_list,area_list)
```
