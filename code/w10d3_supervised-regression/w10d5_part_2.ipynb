## 2. Regression Algorithms - Part 2

### Boston Dataset

```python
import pandas as pd
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split

X,y = load_boston(return_X_y=True)
cols = load_boston()["feature_names"]
X = pd.DataFrame(X, columns=cols)
```

```python
X.head()
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y)
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
y_test[:10]
```

### Basic Linear Regression

```python
# Importamos el modelo desde SKLEARN
from sklearn.linear_model import LinearRegression

# Creamos una instancia del algoritmo que vamos a entrenar, en este momento los coeficientes no estan calculados 
lr = LinearRegression()

# Usando las features y el ground thruth de entrenamiento, calculamos los coeficientes
lr.fit(X_train,y_train)

# Predecimos una vez entrenado el modelo, que tal bien predice los propios datos con los que lo he entrenado
y_pred_train = lr.predict(X_train)

# Predecimos el dataset de testing para validar y mas adelante cuanto error he cometido
y_pred_test = lr.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error as mse

print("Train RMSE:", mse(y_train,y_pred_train)**.5)
print("Test RMSE:", mse(y_test,y_pred_test)**.5)
```

### Feature Crossing
`Feature Crossing` es la creación de una feature sintética multiplicando 2 o más features.

```python
X_cross = X.copy()
```

```python
columns = X_cross.columns
for i in range(len(columns)):
    for j in range(i,len(columns)):
        col_1 = columns[i]
        col_2 = columns[j]
        X_cross[f"{col_1}*{col_2}"] = X_cross[col_1]*X_cross[col_2]
```

```python
X.shape, X_cross.shape
```

```python
X_cross.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_cross,y)
lr = LinearRegression()
lr.fit(X_train,y_train)
y_pred_train = lr.predict(X_train)
y_pred_test = lr.predict(X_test)

print("Train:", mse(y_train,y_pred_train)**.5)
print("Test:", mse(y_test,y_pred_test)**.5)
```

### Standarization

El objetivo de estandarizar es eliminar el factor del error numérico en el entrenamiento de los algoritmos


`StandardScaler()`: Normaliza los datos para que tengan media =0 y desviación típica =1

IMPORTANTE: Solo estandarizamos variables numericas, no las categóricas

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_norm = X.copy()

# Sólo las numéricas
to_normalize = ['CRIM', 'ZN', 'INDUS', 'NOX', 'RM', 'AGE', 'DIS', 'TAX','PTRATIO', 'B', 'LSTAT']
X_norm[to_normalize] = sc.fit_transform(X[to_normalize])
```

```python
X_norm.head()
```

```python
X_norm.describe()
```

```python
import numpy as np

col = np.arange(10)
u = np.mean(col)

col_1 = col - u
col_1

std = np.std(col_1)

col_2 = col_1 / std

np.std(col_2)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_norm,y)
lr = LinearRegression()
lr.fit(X_train,y_train)
y_pred_train = lr.predict(X_train)
y_pred_test = lr.predict(X_test)

print("Train:", mse(y_train,y_pred_train)**.5)
print("Test:", mse(y_test,y_pred_test)**.5)
```

### With Feature Crossing

- Selecting only columns with high correlation to target

```python
X_cross = X_norm.copy()
columns = X_cross.columns
for i in range(len(columns)):
    for j in range(i,len(columns)):
        col_1 = columns[i]
        col_2 = columns[j]
        X_cross[f"{col_1}*{col_2}"] = X_cross[col_1]*X_cross[col_2]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_cross,y, random_state=42)
lr = LinearRegression()
lr.fit(X_train,y_train)
y_pred_train = lr.predict(X_train)
y_pred_test = lr.predict(X_test)

print("Train:", mse(y_train,y_pred_train)**.5)
print("Test:", mse(y_test,y_pred_test)**.5)
```
