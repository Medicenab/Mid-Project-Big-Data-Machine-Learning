```python
import requests
```

```python
BCN=[]
for i in range(0,340):
    BCN.append(requests.get(f"https://api-midproject.herokuapp.com/BCN/{i}").json())
```

```python
Ciutat_Vella_tot=[]
for x in BCN:
    if x['District Name']=="Ciutat Vella":
        Ciutat_Vella_tot.append(x['Number'])
Ciutat_Vella_tot=sum(Ciutat_Vella_tot)

Ciutat_Vella_nat=[]
for x in BCN:
    if x['District Name']=="Ciutat Vella":
        if x['Nationality']=="Spain":
            Ciutat_Vella_nat.append(x['Number'])
Ciutat_Vella_nat=sum(Ciutat_Vella_nat)

Ciutat_Vella_ex=[]
for x in BCN:
    if x['District Name']=="Ciutat Vella":
        if x['Nationality']!="Spain":
            Ciutat_Vella_ex.append(x['Number'])
Ciutat_Vella_ex=sum(Ciutat_Vella_ex)    
        
```

```python

Ciutat_Vella={}
poblacion=[]
for x in BCN:
    if x['District Name']=="Ciutat Vella":
        if x['Nationality']==x['Nationality']:
            poblacion.append(x['Number'])   
            Ciutat_Vella[x['Nationality']]=(poblacion)
```

```python
Ciutat_Vella={}
poblacion=[]
pobla=[]
for x in BCN:
    if x['District Name']=="Ciutat Vella":
        if x['Nationality']=="Spain":
            poblacion.append(x['Number'])

        else:
            pobla.append(x['Number'])
            Ciutat_Vella["extranjero"]=(pobla)
poblacion=sum(poblacion)
pobla=sum(pobla)
Ciutat_Vella["extranjero"]=(pobla)
Ciutat_Vella["Spain"]=(poblacion)
```
